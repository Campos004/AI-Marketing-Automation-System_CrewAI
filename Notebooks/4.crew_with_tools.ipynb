```python
from dotenv import load_dotenv
import os
load_dotenv()
```

```python
from crewai import LLM

llm=LLM(
    model="cerebras/llama-3.3-70b",
    temperature=0.1
)
```

```python
from crewai import Agent,Task,Crew
from crewai_tools import SerperDevTool
from crewai_tools import TavilySearchTool

research_agent=Agent(
    role="Research Specialist",
    goal="Research interesting facts about the topic:{topic}",
    backstory="You are an expert at finding relevent and factual data.",
    tools=[TavilySearchTool()],
    verbose=True,
    llm=llm
)

writer_agent=Agent(
    role="Creative Writer",
    goal="Write a short blog summary using the reasearch",
    backstory="You are skilled at writting engaging summaries based on provided content.",
    verbose=True,
    llm=llm
)
```

```python
task1=Task(
    description="Find 3-5 interesting facts about the topic:{topic} as of year 2026.",
    expected_output="A bullet list of 4-5 facts",
    agent=research_agent,
)

task2=Task(
    description="Write a 120 words blog post summary about {topic} using the facts from the research.",
    expected_output="A blog post summary of 120 words",
    agent=writer_agent,
    context=[task1],
)

crew=Crew(
    agents=[research_agent,writer_agent],
    tasks=[task1,task2],
    verbose=True
)

crew.kickoff(inputs={"topic":"the future of electrical vechicles"})
```
